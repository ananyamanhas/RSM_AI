```python
def binary_encode_column(df, column_name, value_for_zero, value_for_one):
    #mapping
    value_mapping = {value_for_zero: 0, value_for_one: 1}
    df[column_name] = df[column_name].map(value_mapping)
    return df
```

```python
df = binary_encode_column(data, 'Sex', 'female', 'male')
# df = binary_encode_column(data, 'Embarked', 'C', 'S')
df
```

```python
X = df.drop('Survived', axis=1).values 
y = df['Survived'].values  
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=366)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  # Standardize features
X_test = scaler.fit_transform(X_test)  # Standardize features
```

```python
#defining the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
#implementing the logistic regression
def logistic_regression(X, y, alpha=0.01, iters=500):
    #get the number of samples and features
    m, n = X.shape 
    
    #initializing weights and bias to 0
    weights = np.zeros(n)
    bias = 0

    #gradient descent
    for i in range(iters):
        model = np.dot(X, weights) + bias
        
        y_pred = sigmoid(model) #gets the probabilty of the prediction being 1
        
       
        dw = (1/m) * np.dot(X.T, (y_pred - y))
        db = (1/m) * np.sum(y_pred - y)
        
        
        weights -= alpha * dw
        bias -= alpha * db

    return weights, bias
```

```python
#training the model
weights, bias = logistic_regression(X_train, y_train) 
```

```python
def pred(X, weights, bias):
    raw = np.dot(X, weights) + bias 
    #^gets the predictions not restricted between 0 and 1
    y_pred = sigmoid(raw) 
    #^squishes the predictions into the sigmoid functio and you get the probability of it being 0 or 1
    return [1 if i > 0.5 else 0 for i in y_pred]
```

```python

preds = pred(X_test, weights, bias)
```

```python
accuracy = np.mean(preds == y_test) * 100
print(f"Accuracy: {accuracy:.2f}%")
```
